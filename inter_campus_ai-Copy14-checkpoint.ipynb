```python
#join the test and trainset to ensure uniform preprocessing
ntrain = df1.shape[0]
ntest = df2.shape[0]
y = df1[['Promoted_or_Not']]
all_data = pd.concat((df1, df2)).reset_index(drop=True)
all_data.drop(['Promoted_or_Not'], axis=1, inplace=True)
print("all_data size is : {}".format(all_data.shape))
```

```python
# fill missing values with the highest occuring class in the qualification column
all_data['Qualification'] = all_data['Qualification'].fillna(all_data['Qualification'].mode()[0])
```

# a little feature engineering here

```python
all_data['Phd_MBA_MSc']=0
all_data['degree_HND']=0
all_data['non_univ']=0
#df2['tm_tsv']=0
all_data.loc[(all_data['Qualification']=='MSc, MBA and PhD'), 'Phd_MBA_MSc']=1
all_data.loc[(all_data['Qualification']=='First Degree or HND'), 'degree_HND']=1
all_data.loc[(all_data['Qualification']=='Non-University Education'), 'non_univ']=1
```

```python
#all_data.drop('Qualification', axis=1, inplace=True)
```

# assign values to categorical feature classes in the dataset

```python
all_data.replace('Yes', 1, inplace=True)
all_data.replace('No', 0, inplace=True)
all_data.replace('Direct Internal process', -1, inplace=True)
all_data.replace('Agency and others', 0, inplace=True)
all_data.replace('Referral and Special candidates', 1, inplace=True)
all_data.replace('Female', 1, inplace=True)
all_data.replace('Male', -1, inplace=True)
all_data.replace('Married', -1, inplace=True)
all_data.replace('Single', 1, inplace=True)
all_data.replace('Not_Sure', 0, inplace=True)
```

# label encode the remaining categorical features

```python
from sklearn.preprocessing import LabelEncoder
cols = ['Qualification', 'Division',
        'No_of_previous_employers',
        'State_Of_Origin']

# process columns, apply LabelEncoder to categorical features
for c in cols:
    lbl = LabelEncoder() 
    lbl.fit(list(all_data[c].values)) 
    all_data[c] = lbl.transform(list(all_data[c].values))

# shape        
print('Shape all_data: {}'.format(all_data.shape))
```

# drop some unpredictive features

```python
all_data.drop(['State_Of_Origin', 'Channel_of_Recruitment' ,'Year_of_birth','Foreign_schooled',
          'Year_of_recruitment','Gender','Marital_Status',
          'Past_Disciplinary_Action',
         'No_of_previous_employers'], axis=1, inplace=True)

```

# describe to gain knowledge of the skew nature of the dataset

```python
all_data.describe()
```

```python
num_feats=all_data.dtypes[all_data.dtypes!='object'].index
#calculate skew features
skew_feats=all_data[num_feats].skew().sort_values(ascending=False)
skewness=pd.DataFrame({'skew':skew_feats})
skewness
```

```python
#use scipy to control the skew nature of the dataset

skewness = skewness[abs(skewness) > 0.3]
print("There are {} skewed numerical features to Box Cox transform".format(skewness.shape[0]))

from scipy.special import boxcox1p
skewed_features = skewness.index
lam = 0.15
for feat in skewed_features:
    all_data[feat] += 1
    all_data[feat] = boxcox1p(all_data[feat], lam)
    
all_data[skewed_features] = np.log1p(all_data[skewed_features])
```

```python
train = all_data[:ntrain]
test = all_data[ntrain:]
```

```python
x_train,x_test,y_train,y_test=train_test_split(train,y, test_size=0.25)
```

# import and initialize xgboost classifier

```python
import xgboost
from xgboost import XGBClassifier
from sklearn.metrics import f1_score
```

```python
xgb = XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
       colsample_bytree=0.9, gamma=0, learning_rate=0.075, max_delta_step=0,
       max_depth=7, min_child_weight=4, missing=None, n_estimators=400,
       n_jobs=1, nthread=None, objective='binary:logistic', random_state=0,
       reg_alpha=0, reg_lambda=1, scale_pos_weight=1, seed=None,
       silent=True, subsample=1,eval_metric='auc')
```

# train the model

```python
xgb.fit(x_train, y_train)
```

```python
xgb.score(x_test, y_test)
```

# check for f1_score

```python
f1_score(xgb.predict(x_test), y_test)
```
